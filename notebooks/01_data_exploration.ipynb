```python
# Data manipulation
import pandas as pd
import numpy as np
```

```python
# Visualization 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Settings
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')

%matplotlib inline

print("✅ Libraries loaded successfully!")
```

```python
# Load Lending Club data
df = pd.read_csv('../data/accepted_2007_to_2018Q4.csv')

print(f"✅ Data loaded successfully!")
print(f"\nDataset Shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

```python
# Display first few rows
df.head()
```

```python
# See all column names
print("Column Names:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
# Check loan_status distribution
print("Loan Status Values:")
print(df['loan_status'].value_counts())
```

```python
print("\n" + "="*60)
```

```python
# Create binary target variable: 1 = default, 0=paid
default_statuses = ['Charged Off', 'Default',
                    'Late (31-120 days)', 'Late(16-30 days)',
                    'Does not meet the credit policy. Status:Charged Off']

df['default'] = df['loan_status'].apply(
    lambda x:1 if x in default_statuses else 0
)

print("\nDefault Distribution:") 
print(df['default'].value_counts())
print(f"\n✅ Overall Default Rate: {df['default'].mean():.2%}")
```

```python
# Check for missing value
missing = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing_Percentage': missing.values
})

print("Top 15 columns with missing values:")
print(missing_df[missing_df['Missing_Percentage'] > 0].head(15))
```

```python
import os
os.makedirs('../results', exist_ok=True)

# Fico Score Analysis
plt.figure(figsize=(14, 5))

plt.subplot(1,2,1)
plt.hist(df['fico_range_high'].dropna(), bins=50, edgecolor ='black', color ='steelblue')
plt.title('FICO Score Disttribution', fontsize=14, fontweight='bold')
plt.xlabel('FICO Score')
plt.ylabel('Number of Loans')
plt.grid(axis='y', alpha=0.3)

plt.subplot(1,2,2)
fico_bins = pd.cut(df['fico_range_high'], bins=[600,660,700,740,780, 850])
default_by_fico = df.groupby(fico_bins)['default'].mean()
default_by_fico.plot(kind ='bar', color='coral')
plt.title('Default Rate by FICO Score Range', fontsize=14, fontweight='bold')
plt.ylabel('Default Rate')
plt.xlabel('FICO Range')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../results/default_by_fico.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nDefault Rate by FICO Score:")
print(default_by_fico)
           
```

```python
# Default rate by loan grade
plt.figure(figsize=(10, 6))
grade_default=df.groupby('grade')['default'].mean().sort_index()
grade_default.plot(kind='bar', color='#2ecc71', edgecolor='black')
plt.title('Default Rate by Loan Grade(A=Best, G=Worst)', fontsize=14, fontweight='bold')
plt.xlabel('Loan Grade')
plt.ylabel('Default Rate')
plt.axhline(df['default'].mean(), color='red', linestyle='--', linewidth=2, label='Overall Default Rate')
plt.legend()
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../results/default_by_grade.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nDefault Rate by Loan Grade:")
for grade, rate in grade_default.items():
    print(f"Grade{grade}: {rate:.2%}")
```

```python
# Debt-to-Income ratio analysis
plt.figure(figsize=(10, 6))
dti_bins = pd.cut(df['dti'], bins=[0,10,20,30,40,100])
default_by_dti =df.groupby(dti_bins)['default'].mean()
default_by_dti.plot(kind='bar', color='#e74c3c', edgecolor='black')
plt.title('Default Rate by Debt-to-Income Ratio', fontsize=14, fontweight='bold')
plt.xlabel('DTI Ratio(%)')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../results/default_by_dti.png', dpi=300, bbox_inches='tight')
plt.show()

print('\nDefault Rate by DTI:')
print(default_by_dti)
```

```python
# Defualt rate by loan purpose
plt.figure(figsize=(12, 6))
purpose_default = df.groupby('purpose', observed=False)['default'].mean().sort_values(ascending=False)
purpose_default.plot(kind='barh', color='purple', edgecolor='black')
plt.title('Default Rate by Loan Purpose', fontsize=14, fontweight='bold')
plt.xlabel('Default Rate')
plt.ylabel('Loan Purpose')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('../results/default_by_purpose.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("Default RATE BY LOAN PURPOSE")
print("="*60)
for purpose, rate in purpose_default.items():
    print(f"{purpose:30s}: {rate:.2%}")
```

```python
# Convert issue date to datetime and extract year
df['issue_date'] = pd.to_datetime(df['issue_d'])
df['issue_year'] = df['issue_date'].dt.year

# Calculate default rate by year
yearly_default = df.groupby('issue_year', observed=False)['default'].mean()

# Create visualization
plt.figure(figsize=(14, 7))
yearly_default.plot(kind='line', marker='o', linewidth=3, markersize=10, color='darkblue')
plt.title('Default Rate Over Time (2007-2018) - Financial Crisis Impact', fontsize=16, fontweight ='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Default Rate', fontsize=12)

# Highlight the 2008-2009 financial crisis period
plt.axvspan(2008, 2009, alpha=0.3, color ='red', label='2008-2009 Financial Crisis')
plt.axhline(yearly_default.mean(), color='green', linestyle='--', linewidth=2, label='Average Default Rate')

plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/default_over_time.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("DEFAULT RATE BY YEAR")
print("="*60)
for year, rate in yearly_default.items():
    crisis_marker = " ⚠️ CRISIS" if year in [2008, 2009] else ""
    print(f"{year}: {rate:.2%}{crisis_marker}")
print(f"\nAverage across all years: {yearly_default.mean():.2%}")
```

```python
# Default rate by loan amount
amount_bins = pd.cut(df['loan_amnt'], bins = [0, 5000, 10000, 15000, 20000, 400000])
default_by_amount = df.groupby(amount_bins, observed=False)['default'].mean()

plt.figure(figsize =(10, 6))
default_by_amount.plot(kind='bar', color='green', edgecolor='black')
plt.title('Default Rate by Loan Amount', fontsize=14, fontweight='bold')
plt.xlabel('Loan Amount Range ($)')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../results/default_by_amount.png', dpi=300, bbox_inches='tight')

print("\n" + "="*60)
print("DEFAULT RATE BY LOAN AMOUNT")
print("=" * 60)
for amount_range, rate in default_by_amount.items():
    print(f"$ {amount_range}: {rate:.2%}")
```
